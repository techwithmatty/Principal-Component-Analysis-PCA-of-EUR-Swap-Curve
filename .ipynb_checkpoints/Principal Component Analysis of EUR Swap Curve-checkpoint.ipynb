# Principal Component Analysis (PCA) of EUR Swap Curve

## What is PCA?

Principal Component Analysis (PCA) is a mathematical technique used to **simplify complex datasets** while retaining as much important information as possible. It does this by identifying patterns in the data and transforming it into a new set of variables, called **principal components**, which are **uncorrelated** and **ranked by the amount of variation** they capture from the original data.

## Simple Breakdown of PCA

1. **Goal** 

- The main goal is to reduce the number of variables (or dimensions) in a dataset while preserving the key patterns and structures in the data. 

2. **How PCA works**
- PCA starts by finding directions in the data where the **variation** (or spread) is the **greatest**. These directions are called **principal components**.
- The **first principal component** captures the **maximum possible variation** in the data. The second component, which is **orthogonal** (at right angles) to the first, captures the **next highest variation**, and so on.
3. **Benefits**
- It helps in **visualizing high-dimensional data** by reducing it to 2 or 3 dimensions.
- It **removes redundant information**, as **correlated variables** are combined into **single principal components**.
- It can **improve the performance of machine learning models** by reducing noise and focusing on the most informative features.

## Application of PCA to EUR Swap Curve

Applying Principal Component Analysis (PCA) to the **EUR swap curve** involves analyzing the **changes in interest rates (yields) across different maturities** to identify the main **patterns or factors** driving those changes. This is a common technique in financial risk management, as it helps reduce the dimensionality of the data and interpret the dynamics of the yield curve.

## EUR Swap Daily Absolute Returns (in bps)

```python
# Calculate daily returns
daily_returns = (eur_estr_df - eur_estr_df.shift(1))*100

# Remove erroneous values
daily_returns.replace([np.inf, -np.inf], np.nan, inplace=True)
returns = daily_returns.dropna(axis=0)
```

The graph clearly shows that absolute returns are **highly correlated across maturities** and exhibit a fairly **stationary** pattern, indicating that the data is well-suited for principal component analysis.

## Produce a Covariance Matrix  

The choice of **covariance** or **correlation** matrix for Principal Component Analysis (PCA) depends on the context, but the covariance matrix is often considered better in this scenario as it retains the original scale of the data, which is important especially as the **magnitude of variance** (volatility) is a key part of the analysis.

PCA aims to **identify directions** (principal components) that **maximise variance** in the data. The covariance matrix allows PCA to focus on the **actual variances** and covariances as they **appear in the dataset**, **preserving the natural units** of measurement.

The **correlation** matrix **standardises variables** to have unit variance. While this is useful when variables have different units or scales, it can **obscure the true magnitude of variance** if the scales or units themselves carry important information.

```python
cov_matrix = daily_returns.cov()
cov_matrix
```

## Apply PCA to Covariance Matrix

```python
pca = PCA()
pca.fit_transform(cov_matrix)
```

```python
pca_vars = np.round(pca.explained_variance_ratio_*100,decimals=2)
pca_labels = ['PC'+str(x) for x in range(1,len(pca_vars)+1)]
pca_data = pd.DataFrame(pca_vars,index=pca_labels) 
```

```python
ax = pca_data[:8].plot(kind='bar', figsize=(16, 10), legend=False, fontsize=14)
ax.set_xlabel("Principal Component", fontsize=16)
ax.set_ylabel("% Percentage of Explained Variance", fontsize=16)
plt.title("% of Explained Variance by Principal Components", fontsize=16)
plt.show()
```

- Principal Component Analysis (PCA) **measures movements** within a specific market and represents them as a **combination of a small number of underlying drivers**, known as principal components (PCs).

- In the context of yield curve analysis, changes can typically be described using **three** key factors: **level, slope, and curvature**, corresponding to the first, second, and third principal components.

- PCA formalises this framework and enables us to assess whether particular segments of the yield curve have **cheapened or richened** beyond what would be expected based on recent yield movements.

### Key Point
- The **first three** principal components usually explain **more than 90% of total variation**, so our analysis will focus exclusively on these components.
- Applying PCA decomposes market movements into **uncorrelated and interpretable** factors, making them easier to analsze.
- This process reduces the dimensionality of the original dataset, allowing us to express market views simply by taking **positions on individual factors**.
- As a result, we can **design trades or portfolios** that are either exposed to, or hedged against, specific factors.

## Methodology

- The full information in a dataset can be separated into **structural movements**, captured by the principal components, and **residual noise**.
- If the **residual noise deviates meaningfully from zero**, it may indicate a **structural dislocation** or anomaly within the market.
- These **signals** can then be **evaluated alongside broader market views** to determine whether a viable trading opportunity exists.

What we have calculated up to this point reflects **movements in the principal components** rather than the components themselves. In formal mathematical language, these quantities correspond to **eigenvectors**, though in practice they are typically called **factor loadings**.

These factor loadings, defined for each tenor, will now be stored in a separate DataFrame.

```python
f_loadings = pd.DataFrame({'PC1':pca.components_[0],'PC2':pca.components_[1],'PC3':pca.components_[2]}, index=cov_matrix.index)
f_loadings
```

## Intepret the Curve (using plotted PCs) 

```python
plt.figure(figsize=(8, 5))

for col in f_loadings.columns:
    plt.plot(
        f_loadings.index,
        f_loadings[col],
        marker='o',
        label=col
    )

plt.title("PCs across Tenors")
plt.xlabel("Tenor")
plt.ylabel("Change in Yield")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Given:
- **PC1** drives **parallel shifts** in the curve (level).
- **PC2** redistributes **yield changes** between the **front and back end** (slope).
- **PC3** isolates **relative movements** between the **belly and the wings** (curvature).

### PC1

**PC1** remains **positive across all tenors**, with a **gradual decline from short to long maturities**.

This indicates that shocks to the first principal component result in rates moving in the **same direction across the curve**, consistent with a level shift. The **larger loadings at the front end** imply that **shorter maturities** are **more sensitive** to level movements than **longer maturities**.

### PC2

**PC2** shows a **single change in sign**, transitioning **from positive at the short end to negative at longer maturities**. 

This implies that short-dated and long-dated rates **move in opposite directions**, capturing changes in the slope of the yield curve. A **positive realisation** of PC2 would therefore correspond to a **steepening**, while a **negative realisation** would indicate **flattening**.

### PC3

**PC3** exhibits **two sign changes**, with **positive** loadings at the **short and long ends** and **negative** loadings in the **intermediate maturities**. 

This pattern implies that the **wings of the curve** move **together**, while the **belly** moves in the **opposite direction**. As such, the third principal component captures **changes in curvature**, such as **butterfly-type movements**.

## Compute PCA Values
- Up to now, we have focused on how the principal components affect the curve, rather than their **actual numerical values**.
- The next step is to compute the value of each principal component at each point in time.
- This is straightforward: each principal component is obtained by **combining the original data** with its corresponding **factor loadings**. You can think of these **loadings as weights**, and the **principal component value** as a **weighted average of yield** changes across maturities.
- In practice, this means **multiplying the data by the loadings and adding the results together**.
- **Repeating** this calculation for every date gives us a **time series** for each principal component.

```python
o_data = returns.copy(deep=True)
pcas = np.dot(o_data,f_loadings)

pca_df = pd.DataFrame(pcas,columns=['PC1','PC2','PC3'], index=o_data.index)

# Combining the dataframes -- change in swap rates + PCA (will allow us to plot the data easily)
o_data = o_data.join(pca_df)
pca_df
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 1) Ensure the index is datetime
o_data = o_data.copy()
o_data.index = pd.to_datetime(o_data.index, errors="coerce")

# (optional) drop rows where the date couldn't be parsed
o_data = o_data.loc[o_data.index.notna()]

fig, ax = plt.subplots(figsize=(10, 6))

for col in o_data.columns[-3:]:
    ax.plot(o_data.index, o_data[col], label=col, linewidth=1)

ax.set_title("Actual Values of Principal Components")
ax.set_xlabel("Time")
ax.set_ylabel("Values")

# 2) Cleaner date ticks
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
fig.autofmt_xdate()  # rotates/aligns dates nicely

ax.legend()
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

## Calculating the Expected Daily Change in Swap Rates (as Implied by PCA)

- PCA allows us to **express movements** in the original dataset using a **small set of underlying factors** rather than individual tenors.
- For the swap curve, this means that the **daily change in yield at any given maturity** can be explained by a **combination of three principal components**.
- As an illustration, the daily move in the **10-year swap rate** can be reconstructed by taking the corresponding **factor loadings** (stored in the 'f_loadings' DataFrame) and combining them with the **realised principal component values** (contained in pca_df).
- This reconstruction is **applied consistently** across **all maturities and all dates**.
- From a computational standpoint, the full set of implied daily changes is obtained by taking the **dot product** of **pca_df** with the **transpose of f_loadings**, where the transpose represents the eigenvector matrix aligned by tenor.

```python
exp_change = np.dot(pca_df,f_loadings.T) 
exp_changes = pd.DataFrame(exp_change,index = pca_df.index, columns=returns.columns)
exp_changes
```

## Calculating PCA Residuals and Identifying Relative Value Opportunities

- Once the PCA-implied rates have been computed, the remaining **unexplained component** can be isolated.
- This **residual** is obtained by **subtracting** the rate **implied by the PCA model** from the **observed market rate**.
- **Visualising these residuals** across **tenors** and over time **helps highlight deviations** from **typical curve behaviour**.
- Such deviations can point to **relative value opportunities**, where specific maturities appear **cheap or rich** compared to what the **PCA framework** would suggest.

```python
resid = returns - exp_changes
resid
```

```python
# get last date
last_index = resid.index[-1]

# extract residuals for that date
residuals = resid.loc[last_index]

plt.figure(figsize=(9, 7))
plt.bar(residuals.index, residuals.values)

plt.title(f"PCA Residuals on {last_index}")
plt.xlabel("Tenors")
plt.ylabel("Residual (in bps)")

plt.xticks(rotation=45)
plt.grid(True, axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

## What the PCA residuals imply on 29 Nov 2024 (EUR swaps)

This chart shows, for each tenor, the difference between the observed EUR swap rate and the **rate implied by the PCA model** on that day.

- **Positive residual** â†’ the tenor is **rich** (trading above what the curve structure would suggest)
- **Negative residual** â†’ the tenor is **cheap** (trading below fair value implied by PCA)
- Residuals near zero â†’ broadly **in line with the curve**

Because the PCA model captures the dominant **level, slope, and curvature** dynamics, the residuals represent **idiosyncratic mispricing**, not explained by normal curve moves.

## RV signals on this day

From the residual pattern, as implied by the setup:

- Tenors with the **largest positive residuals** appear **rich** versus the rest of the curve
   â†’ candidates to **receive less / pay more** in RV terms
- Tenors with the **most negative residuals** appear **cheap**
    â†’ candidates to **receive** relative to neighboring maturities
- If the residuals show a **localised hump or dip** (e.g. belly cheap vs wings), this suggests a **butterfly-style RV opportunity**
- If short and long ends show opposite signs, this can indicate a **curve segmentation** not captured by slope alone

Importantly, these are **relative**, not directional, signals:
the trade idea is to **express convergence back toward the PCA-implied curve**, not to bet on outright rate direction.
